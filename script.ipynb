```python
import matplotlib.pylab as plt
import numpy as np
import math
#x1 is the value of the first guess
#this allows the program to ask for an appropriate value from the user until a right value is provided
#the program will give the error message for each inappropriate value guessed by the user
x1=float(input("Enter the value for x1: "))
if x1>=0:
    while x1>=0:
        print("Please enter a correct value for x1! (x1<0)")
        x1=float(input("Enter the value for x1: "))
        
#x3 is the value of the second guess
#this allows the program to ask for an appropriate value from the user until a right value is provided
#the program will give the error message for each inappropriate value guessed by the user
x3=float(input("Enter the value for x3: "))
if x3<=0:
    while x3<=0:
        print("Please enter a correct value for x3 (x3>0)")
        x3=float(input("Enter the value for x3: "))

#determining the value for x2 as instructed in the assignment 
x2=(x1+x3)/2
#checking if f(x) is greater than 0 and changing x2 accordingly
if a*x2*x2+b*x2+c>0:
    x2=x1
#what to do if f(x) is less than 0 and changing x2 accordingly
else:
    x2=x3

tol=0.0001   #------------------initializing the tolarence value

point_f= a*x2*x2+b*x2+c   #-----determining the value for f(x) for each x2
nsteps=0    #-------------------counting the number of tests
#loop for updating the value of x2 and putting it in f(x) until value for f(x) is less than the tolerance value
while abs(point_f)>tol:
    if point_f>0:
        x1=x2
    else:
        x3=x2
    x2=(x1+x3)/2
    point_f= a*x2*x2+b*x2+c
    nsteps=nsteps+1
#printing the required values as instructed
print("The value new x1 and x3 are %.5f and %.5f respectively." %(x1,x3))
print("The value for x2 and f(x2) are %.5f and %.5f respectively." %(x2, point_f))
print("The number of steps taken to determine the root is ",nsteps)
x=np.arange(-4.0,6.0,0.2)
plt.plot(x,a*x*x+b*x+c)
plt.plot(x,0.0*x)
plt.plot(x2,a*x2*x2+b*x2+c,'ro')
plt.show()
```
